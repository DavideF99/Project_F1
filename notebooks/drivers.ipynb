```python
import requests

url = "https://api.openf1.org/v1/drivers"

response = requests.get(url)
```

```python
print(response.status_code)
```

```python
driver_data = response.json()
driver_data
```

```python
driver_data[0]
```

```python
# Get a list of drivers names using a for loop

drivers_names = []

for driver in driver_data:
    drivers_names.append(driver["full_name"])

drivers_names
```

```python
# Get a list of drivers names using set and list comprehension

driver_names = list({driver['full_name'] for driver in driver_data})
driver_names
```
